```python
import matplotlib.pyplot as plt
import numpy as np
import math
import matplotlib.colors
```

```python
def syracuse(u_n):
    if u_n % 2 == 0 :
        u_n1 = u_n/2
    else : 
        u_n1 = 3*u_n + 1

    return u_n1
```

```python
def test_conjecture (u_0):
    u_n = u_0
    n=0
    n_tab=[0]
    tableau_valeurs = [u_0]

    while u_n != 1 :
        u_n1 = syracuse(u_n)
        u_n=u_n1
        n=n+1
        tableau_valeurs.append(u_n)
        n_tab.append(n)

        maxi = max(tableau_valeurs)
        if type(maxi) == int:
            altitude_u_0=maxi
            
            
    
    u_p=u_n

    return u_p,n, tableau_valeurs, n_tab, altitude_u_0
    
```

```python
u_0 = int(input("u_0=?"))
if u_0 < 1 :
    print("erreur")
else : 
    u_p,n, tableau_valeurs, n_tab, altitude_u_0= test_conjecture(u_0)
    print("u_p=", u_p, "à partir du rang", n, "altitude u_0=", altitude_u_0)
    print(tableau_valeurs)
     
    plt.plot(n_tab,tableau_valeurs, color='mediumaquamarine')
    plt.xlabel('Itération')
    plt.ylabel('Valeur de la suite')
```
